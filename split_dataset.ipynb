```
import os, shutil, numpy

image_dir = '/content/drive/My Drive/Colab Notebooks/no-mask/data/images'
train_dir = '/content/drive/My Drive/Colab Notebooks/no-mask/data/images/train'
val_dir = '/content/drive/My Drive/Colab Notebooks/no-mask/data/images/val'


def count_files(path):
    files = os.listdir(path)
    return len(files)


file_count = (count_files(image_dir))
print(file_count)
```

```
src_files = sorted(os.listdir(image_dir))
# src_files.pop('.ipynb_checkpoints')
n_files = len(src_files)

split_ratio = 0.8
split_index = int(n_files * split_ratio)

# numpy.random.shuffle(src_files)

print(src_files[0:split_index][:6])
print(src_files[split_index:][:6])

train_src = src_files[0:split_index]
val_src = src_files[split_index:]

for i in train_src:
    shutil.copy(os.path.join(image_dir, i), train_dir)

for i in val_src:
    shutil.copy(os.path.join(image_dir, i), val_dir)
```
